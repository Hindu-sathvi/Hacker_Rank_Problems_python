```python
file= open('sample_data/california_housing_test.csv')
content=file.read()
file.close()
```

```python
import csv
data_dict={}
for index,row in enumerate(csv.reader(content.strip().split('\n'))):
  if index ==0:
    for column in row:
      data_dict[column]=[]
  else:
    for col_index,value in enumerate(row):
      data_dict[list(data_dict.keys())[col_index]].append(value)
```

```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
dataset =load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary=df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.xlabel('Salary($USD)')
plt.ylabel('')
plt.title('Median Salary by job Title')
plt.show()
```
